```python
import tensorflow as tf
w=tf.Variable([.3],dtype=tf.float32)
b=tf.Variable([.3],dtype=tf.float32)

## model input and output
x=tf.Variable([5.0],dtype=tf.float32)
y=w*x+b
print(y)
```

```python
## Linear Model in tensorflow
```

```python
train_x=[1,2,3,4]
train_y=[2,4,6,8]
NUM_EXAMPLES=len(train_x)
epochs=1000
## TRAINING INFO
w=tf.Variable(0.0)
b=tf.Variable(0.0)
training_steps=100
learning_rate=0.01
for epoch in range(epochs):
  ## forward pass
  with tf.GradientTape() as g:
    y_pred=w*train_x+b
    loss=tf.reduce_mean(tf.square(y_pred-train_y))
  ##evaluate the gradient with respect to loss
  dw,db=g.gradient(loss,[w,b])
  ## update the parameters
  w.assign_sub(learning_rate*dw)
  b.assign_sub(learning_rate*db)
  ##print the loss for every 20 iterations
  if epochs%20==0:
    print('loss at step {:03d}: {:0.3f}'.format(epochs, loss))
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(train_x,train_y,'ro',label='data')
plt.plot(train_x,np.array(w*train_x+b),label='model')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=wines.drop(['type'],axis=1)
y=wines['type']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
## DeepNeural Network
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
scaler=StandardScaler().fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
model=Sequential()
## add an input layer
model.add(Dense(10,activation='relu',input_shape=(5,)))
## add hidden layer
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,epochs=10,validation_split=0.2)
```

```python
loss,acc=model.evaluate(x_test,y_test)
loss,acc=model.evaluate(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import pandas as pd

y_pred = model.predict(x_test)          # continuous probabilities
y_pred_classes = (y_pred > 0.5).astype(int)  # threshold at 0.5

print(pd.DataFrame(confusion_matrix(y_test, y_pred_classes)))
print(classification_report(y_test, y_pred_classes))
```
